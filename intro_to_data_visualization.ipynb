# Bar Plots

Default settings to fancy

```python
#seaborn bar plot with default settings
"""
data = dataset to use
x = IV1
y = DV
hue = IV2
"""
sns.barplot(data=wm_df, x="SetSize", y="Accuracy", hue="Domain")
```

**Things to think about:**

Do you like the axis labels?

What are the error bars?

Do you want the y-axis to start at zero?

```python
#seaborn plot with customizations
sns.barplot(data=wm_df, x="SetSize", y="Accuracy", hue="Domain", palette='colorblind',
                capsize=0.1, err_kws={'linewidth': 2}) #length error caps, thickness error bars

#customize axes labels
plt.ylabel("Proportion Correct")
plt.xlabel("Set Size")
plt.legend(title="Sensory Domain")

#change the start of y-axis
plt.ylim(0.45, 1.0) #jp hack: leave room for an axis breaker

#remove the border
sns.despine()

#show plot
plt.tight_layout()
plt.show()
```

```python
#let's get fancy - adding individual data points overlaid

#first plot - bar plot
sns.barplot(data=wm_df, x="SetSize", y="Accuracy", hue="Domain", palette='colorblind',
                capsize=0.1, err_kws={'linewidth': 2}) #length error caps, thickness error bars

#second plot - strip plot
sns.stripplot(data=wm_df, x="SetSize",y="Accuracy",hue="Domain",dodge=True,
              palette='dark:black',legend=False)

#customize axes labels
plt.ylabel("Proportion Correct")
plt.xlabel("Set Size")
plt.legend(title="Sensory Domain")

#change the start of y-axis
plt.ylim(0.45, 1.0) #jp hack: leave room for an axis breaker

#remove the border
sns.despine()

#show plot
plt.tight_layout()
plt.show()
```

**Things to think about:**

Now we can't see the error bars

The individual data points are all on top of each other

```python
#adding jitter and changing alpha

#first plot - bar plot
sns.barplot(data=wm_df, x="SetSize", y="Accuracy", hue="Domain", palette='colorblind',
                capsize=0.1, err_kws={'linewidth': 2}) #length error caps, thickness error bars

#second plot - strip plot
sns.stripplot(data=wm_df, x="SetSize",y="Accuracy",hue="Domain",dodge=True,
              palette='dark:black',legend=False,
              jitter = .25, alpha=.4) #jitter and alpha change - can play with these values

#customize axes labels
plt.ylabel("Proportion Correct")
plt.xlabel("Set Size")
plt.legend(title="Sensory Domain")

#change the start of y-axis
plt.ylim(0.45, 1.0) #jp hack: leave room for an axis breaker

#remove the border
sns.despine()

#show plot
plt.tight_layout()
plt.show()
```
